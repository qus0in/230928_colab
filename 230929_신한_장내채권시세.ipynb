```python
import requests
import pandas as pd
import plotly.express as px

class bond:
    def __init__(self):
        self.get_prices()
        pass

    def get_prices(self):
        # 데이터 받기
        URL = 'https://www.shinhansec.com/siw/wealth-management/bond-rp/590401/data.do'
        response = requests.get(URL)
        data = response.json().get('body').get('반복데이타0')
        # 테이블 전처리
        df = pd.DataFrame(data)
        df.columns = [
            '종목명', '매도수익률', '잔존만기일수', '거래대금', '표면금리',
            '신용도', '종목코드', '현재가', '발행주체', '이자지급방법',
            '잔존만기_연', '잔존만기_월', '잔존만기_일', '매수수익률', '표면금리2',
            'x1', 'x2','만기일시', '거래량', '등락률', '전일종가',
        ]
        # 필드 전처리
        df.종목명 = df.종목명.str.strip()
        df.잔존만기일수 = df.잔존만기일수.astype(int)
        df.거래량 = df.거래량.astype(int)
        df.표면금리 = df.표면금리.astype(float)
        df.매수수익률 = df.매수수익률.astype(float)
        # 신용도
        국공채 = lambda x: 'AA' if (x.신용도 == ' ' and x.발행주체 != '회사채') else x.신용도
        빈칸 = lambda x: x.신용도 if x.신용도 != ' ' else self.get_credit(x.종목코드)
        df.신용도 = df.apply(국공채, axis=1)
        df.신용도 = df.apply(빈칸, axis=1)
        # 세후매수수익률
        df['세후매수수익률'] = df.매수수익률.mul(1 - 0.154).apply(lambda x: int(x * 100) / 100)
        self.prices = df.iloc[:, [0, 2, 18, 4, 5, 6, 8, 13, 21]]\
                .set_index('종목코드')\
                .sort_values('거래량', ascending=False)

    @classmethod
    def get_credit(cls, symbol):
        URL = 'https://www.shinhansec.com/siw/wealth-management/bond-rp/590401P02/data.do'
        response = requests.post(URL, json=dict(
            bondCode=symbol, cls=symbol,
        ))
        return response.json().get('body').get('bondMaster')['신용등급명']

    @classmethod
    def draw_chart(cls, data, title):
        fig = px.scatter(
            data,
            x = '매수수익률', y ='잔존만기일수', log_y = True,
            text = '종목명', color = '신용도',
            size = '거래량', size_max = 50,
            title = title, height = 525,
        )
        fig.show()

b = bond()
```

```python
top = b.prices.head(20)
print(top.iloc[:, [0, 5, 4, 6, 7]]
      .to_markdown(index=False))
b.draw_chart(top, '거래량 TOP20')
```

```python
def draw_chart(data, title):
    fig = px.scatter(
        data,
        x = '매수수익률', y ='잔존만기일수', log_y = True,
        text = '종목명', color = '신용도',
        size = '거래량', size_max = 50,
        title = title, height = 525,
    )
    fig.show()

draw_chart(top, '거래량 TOP20')
```

```python
#@title 거래량 평균 이상, 투자등급(BBB- 이상), 잔존 만기 1~5년 회사채 매수수익률 TOP15
screen_top = df.loc[df.발행주체 == '회사채']\
    .loc[df.거래량 > df.거래량.mean()]\
    .loc[df.잔존만기일수 > 365]\
    .loc[df.잔존만기일수 <= (365 * 5)]\
    .query('(not 신용도.str.contains("B")) or 신용도.str.contains("BBB")')\
    .sort_values('매수수익률', ascending=False)\
    .head(20)

screen_top.iloc[:, [0, 4, 6, 7]]
```

```python
draw_chart(
    screen_top,
    '거래량 평균 이상, 투자등급(BBB- 이상), 잔존 만기 1~5년 회사채 매수수익률 TOP15')
```
